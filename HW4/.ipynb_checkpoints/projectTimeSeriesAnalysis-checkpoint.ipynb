```python
import numpy as np
import pandas as pd

import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns

import os
from datetime import datetime

pd.options.display.max_columns = None
```

```python
path = './project_input/activities_with_zones.csv'
df = pd.read_csv(path)
df.head()
```

```python
dist_df = df[['start_date_local', 'distance']]
dist_df.head()
```

```python
dist_df.columns = ['date', 'distance']
dist_df.head()
form_dist_df = pd.DataFrame({'date': [], 'distance': []})
```

```python
form_dist_df['date'] = pd.to_datetime(dist_df['date'])
form_dist_df['date'].iloc[0]
```

```python
form_dist_df['distance'] = dist_df['distance'].apply(lambda x : x / 1000)
form_dist_df['distance'].iloc[0]
```

```python
form_dist_df.head()
```

```python
form_dist_df.tail()
```

```python
plt.figure(figsize=(15,4), dpi=100)
plt.plot(form_dist_df['date'], form_dist_df['distance'], color='tab:red')
plt.gca().set(title='Distance per ride from 2013 to 2023', xlabel='Date', ylabel='Distance')
plt.show()
```

```python
form_dist_df.describe()
```

```python
df_2023 = form_dist_df[form_dist_df['date'].dt.year == 2023]
df_2023.describe()
```

```python
plt.figure(figsize=(15,4), dpi=100)
plt.plot(df_2023['date'], df_2023['distance'], color='tab:orange')
plt.gca().set(title='Distance per ride from 2023', xlabel='Date', ylabel='Distance')
plt.show()
```

```python
df_weekly_23 = df_2023.copy()
df_weekly_23['week'] = df_2023['date'].dt.isocalendar().week
df_grouped_weekly = df_weekly_23.groupby('week')
df_aggregated = df_grouped_weekly['distance'].sum().to_frame()
df_aggregated['week'] = df_aggregated.index
df_aggregated.head()
```

```python
plt.figure(figsize=(15,4), dpi=100)
plt.plot(df_aggregated['week'], df_aggregated['distance'])
plt.plot(df_aggregated['week'], df_aggregated['distance'], marker='o', linestyle=None)
plt.gca().set(title='Distance per week from 2023', xlabel='Date', ylabel='Distance')
plt.show()
```
